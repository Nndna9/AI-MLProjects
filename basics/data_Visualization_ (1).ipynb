Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK

3.Scatter plot

When it comes to scatter plot you can use alot of parameters as you like here are some :
Parameters:

1. x_axis_data: An array containing data for the x-axis.matplotlib
2. s: Marker size, which can be a scalar or an array of size equal to the size of x or y.
3. c: Color of the sequence of colors for markers.
4. marker: Marker style.
5. cmap: Colormap name.
6. linewidths: Width of the marker border.
7. edgecolor: Marker border color.
8. alpha: Blending value, ranging between 0 (transparent) and 1 (opaque).


Here'll I'll just use colour.


```python
x = [5,7,8,3,5,1,6,7,4,3]
y = [99,87,67,58,97,65,77,67,89,98]
plt.scatter(x,y,c="red")
plt.show()
```

4. With multiple values

   For making my plot good and for easy recognition I've used 2 markers "d" and "x" you can also use "s" for square, "^" triangle,etc.

```python
x1 = [1,2,3,4,5,7,7,7]
y1 = [60,77,78,93,76,92,78,61]
```

```python
x2 = [2,2,6,4,4,7,7,8]
y2 = [66,70,78,99,86,62,92,82]
```

```python
plt.scatter(x1,y1, c= "red" , linewidths = 4, marker = "d" )
plt.scatter(x2,y2, c= "yellow" , linewidths = 3, marker = "x" )
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

5. Bubble chart

A bubble chart is a data visualization which helps to displays multiple circles (bubbles) in a two-dimensional plot as same in scatter plot. A bubble chart is primarily used to depict and show relationships between numeric variables.


Similar to a random scatter plot you can also make a random bubble plot.

```python

# Data
x_values = [1, 2, 3, 4, 5]
y_values = [2, 3, 5, 7, 11]
bubble_sizes = [50, 100, 150, 200, 300]

# Create a bubble chart with customization
plt.scatter(x_values, y_values, s=bubble_sizes, alpha=0.5, edgecolors='g', linewidths=2)

# Add title and axis labels
plt.title("Bubble Chart with Transparency")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.show()
```

Random

Here we use random() function to get random values,colors and sizes


We've also used a color bar for making it look good.

```python

import numpy as np

# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 100 * np.random.rand(50)

# Create a customized scatter plot
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis')

# Add title and axis labels
plt.title("Customized Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display color intensity scale
plt.colorbar(label='Color Intensity')

# Show the plot
plt.show()
```
