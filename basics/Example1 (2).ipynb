```python
colors=['red','blue']
Quality=['good', 'bad']
for i in range(3):
    x=df[df['Quality']== Quality[i]]
    plt.scatter(x['Size'],x['Weight'],c=colors[i],label=Quality[i])
plt.xlabel("Apple size")
plt.ylabel("Weight")
plt.legend()
```

The IndexError: list index out of range error occurs in Python when an item from a list is attempted to be accessed that is outside the index range of the list. The range of a list in Python is [0, n-1], where n is the number of elements in the list. Here the used data set is too big
